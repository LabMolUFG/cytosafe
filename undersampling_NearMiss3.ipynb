```python
import pycaret
from pycaret.classification import *

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import MACCSkeys

from collections import Counter
from imblearn.under_sampling import NearMiss
from imblearn.under_sampling import ClusterCentroids

from numpy import where
```

```python
df = pd.read_csv('...')
df
```

```python
df.groupby('Outcome').size()
```

```python
df['mol'] = [Chem.MolFromSmiles(x) for x in df['SMILES']]
fps = np.array([AllChem.GetMorganFingerprintAsBitVect(x, radius = 2, nBits = 1024, useFeatures=False) for x in df['mol']])
#fps = np.array([MACCSkeys.GenMACCSKeys(x) for x in df['mol']])
```

```python
X = fps
y = df[['Outcome', 'SMILES', 'CAS No', 'Source' ]]

X = pd.DataFrame(X)
y = pd.DataFrame(y)
```

```python
undersample = NearMiss(sampling_strategy='majority', version=3, n_neighbors_ver3=50)
# transform the dataset
X_resampled, y_resampled = undersample.fit_resample(X, y['Outcome'])
```

```python
index_under = undersample.sample_indices_
index_under
```

```python
novo_df = y.iloc[index_under]
```
